### Merging my booktok merged list AND the kindle list to create BI visuals...

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
booktok_on_kindle = pd.read_csv(r"C:\Users\jenni\OneDrive\Desktop\DA12\Projects\Goodreads_Kindle_NSScapstone\data\CLEANED\booktok_on_kindle.csv")
```

```python
kindle = pd.read_csv(r"C:\Users\jenni\OneDrive\Desktop\DA12\Projects\Goodreads_Kindle_NSScapstone\data\CLEANED\final_kindle_books.csv")
```

```python
booktok_on_kindle['Booktok'] = 'Y'
```

```python
kindle['Booktok'] = 'N'
```

```python
merged_all = pd.merge(booktok_on_kindle, kindle, on='key', how='outer')
merged_all.head()
```
