```python
#grab the urls for scraping
urls = booktok_books['Goodreads URL'].tolist()
print(urls)
```

```python
len(urls) #check to make sure all are there 
```

```python
# Lists to hold the scraped data
book_titles = []
book_reviews = []

# Iterate over each URL to scrape book info
for url in urls:
    try:
        response = requests.get(url)  # Use requests to get the book page
        if response.status_code == 200:
            soup = BS(response.content, 'html.parser')  # Parse the page content
            
            # Scrape the book title
            title_tag = soup.find('h1', class_='gr-h1 gr-h1--serif')
            book_title = title_tag.get_text(strip=True) if title_tag else "Title not found"
            book_titles.append(book_title)
            
            # Scrape the book reviews
            reviews = soup.find_all('div', class_='reviewText stacked')
            book_reviews_for_this_book = [review.get_text(strip=True) for review in reviews]
            book_reviews.append(book_reviews_for_this_book)  # List of reviews for this book
            
        else:
            print(f"Failed to retrieve details for {url}: {response.status_code}")
    except Exception as e:
        print(f"An error occurred while processing {url}: {e}")

    # Add a random delay to avoid rate limiting
    time.sleep(10)  # Random delay between 2 to 5 seconds
    
# Convert lists to DataFrame for easier handling
# results = pd.DataFrame({
#     'Book Title': book_titles,
#     'Reviews': book_reviews
# })

# # Output the scraped data
# print(results)
```
