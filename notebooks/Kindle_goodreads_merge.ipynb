```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
booktok_books = pd.read_csv(r'C:\Users\jenni\OneDrive\Desktop\DA12\Projects\Goodreads_Kindle_NSScapstone\notebooks\booktok_maybeclean1.csv')
```

```python
booktok_books.head()
```

```python
#drop the format column, don't need

booktok_books = booktok_books.drop('format', axis=1)
booktok_books.head()
```

```python
# rename columns for later merge

booktok_books = booktok_books.rename(columns={'avg rating': 'Goodreads Avg Rating', 'num_ratings': 'Number of Goodreads Ratings', 'genre': 'Goodreads Top Genre', 'num_pages': 'Number of Pages'})
booktok_books.head()
```

```python
kindle_books = pd.read_csv(r"C:\Users\jenni\OneDrive\Desktop\DA12\Projects\Goodreads_Kindle_NSScapstone\data\kindle_maybeclean1 - USECSV.csv")
```

```python
kindle_books.head()
```

```python
kindle_books.shape
```

```python
#drop a few more columns

kindle_books = kindle_books.drop(['soldBy', 'imgUrl', 'publishedDate'], axis=1)
kindle_books.head()
```

```python
#rename a few columns for clarity in the merge later 

kindle_books = kindle_books.rename(columns={'productURL': 'Amazon URL', 'stars': 'Kindle Avg Rating', 'reviews': 'Number of Kindle Reviews', 'price': 'Kindle Price'})
kindle_books.head()
```

## key column now: booktok

```python
booktok_books['key'] = booktok_books['Title'] + booktok_books['Author']
booktok_books.head()
```

```python
#strip whitespace from the key

booktok_books['key'] = booktok_books['key'].str.replace(r'\s+', '', regex=True)
booktok_books.head()
```

```python
# take out all periods (.)

booktok_books['key'] = booktok_books['key'].str.replace('.', '')
booktok_books.head()
```

```python
# reset the capitals to lower

booktok_books['key'] = booktok_books['key'].str.lower()
booktok_books.head()
```

## now for kindle 

```python
kindle_books['key'] = kindle_books['title'] + kindle_books['author']
kindle_books.head()
```

```python
# strip whitespace from the key

kindle_books['key'] = kindle_books['key'].str.replace(r'\s+', '', regex=True)
kindle_books.head()
```

```python
# take out all periods (.)

kindle_books['key'] = kindle_books['key'].str.replace('.', '')
kindle_books.head()
```

```python
# reset the capitals to lower

kindle_books['key'] = kindle_books['key'].str.lower()
kindle_books.head()
```

## let's try a merge (pray for me)

```python
booktok_on_kindle = pd.merge(booktok_books, kindle_books, on='key', how='inner')
```

```python
booktok_on_kindle.shape
```

```python
booktok_on_kindle
```

```python
#let's drop some of this... 

booktok_on_kindle = booktok_on_kindle.drop(['title', 'author'], axis=1)
booktok_on_kindle.head()
```

```python
#let's reorder these columns to make it easier to compare notes 

booktok_on_kindle = booktok_on_kindle[['key', 'Title', 'Author', 'Kindle Price', 'Goodreads Avg Rating', 'Kindle Avg Rating', 'Number of Goodreads Ratings', 'Number of Kindle Reviews', 'Num_shelved_booktok', 'Goodreads Top Genre', 'category_name', 'year_published', 'Number of Pages', 'isKindleUnlimited', 'isBestSeller', 'isEditorsPick', 'isGoodReadsChoice', 'Goodreads URL', 'Amazon URL', 'Image URL']]
booktok_on_kindle.head()
```

## now let's isolate books that ARE on kindle, but NOT viral on booktok (on the Goodreads list)

```python
not_on_booktok = kindle_books[~kindle_books['key'].isin(booktok_books['key'])]
```

```python
not_on_booktok.head()
```

### Now! Some EDA...

I want to exclude some of the books from the NOT list that don't have comparable categories / genres. It's not a fair comparison to look at textbooks on that list (and their prices and ratings) versus a romance book on the other. Let's see what the top categories are on the merged list and go from there. 

```python
booktok_on_kindle['Goodreads Top Genre'].value_counts()
```

```python
booktok_on_kindle['category_name'].value_counts()
```

Okay so now I have the genres and categories on the booktok list - I need to exclude the ones from the Kindle list that are NOT on here.

```python
exclude_categories = ['Sports & Outdoors', 'Self-Help', 'Reference', 'Medical', 'Law', 'History', 'Health, Fitness & Dieting', 'Engineering & Transportation', 'Education & Teaching', 'Crafts, Hobbies & Home', 'Cookbooks, Food & Wine', 'Computers & Technology', 'Comics', 'Business & Money']

not_on_booktok_filtered = not_on_booktok[~not_on_booktok['category_name'].isin(exclude_categories)]
```
