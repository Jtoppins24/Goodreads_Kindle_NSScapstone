# Notebook goal: webscrape Goodreads Romantasy shelf (page 1)

```python
import requests
from bs4 import BeautifulSoup as BS
import re
import pandas as pd
import numpy as np
import time
from time import sleep
from random import randint
import matplotlib.pyplot as plt
%matplotlib inline
import lxml 
# from selenium import webdriver
# from webdriver_manager.chrome import ChromeDriverManager
```

```python
URL = 'https://www.goodreads.com/shelf/show/romantasy'

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36'}
# used header here to specify that i'm not a bot but a user, that way goodreads let me in
response = requests.get(URL, headers=headers)
print(response.content.decode())
```

```python
response.status_code
```

```python
soup = BS(response.text)
```

```python
type(soup)
```

## get book info from first page

```python
test = soup.findAll('div', {'class':'left'}) #test contains ONLY the book elements and NOTHING ELSE 
print(test)
```

```python
# cards = soup.findAll('div', {'class': 'elementList'}) #cards contains all the elements for each book that i want 
# print(cards)
```

### titles from first page

```python
all_bookTitles = soup.find_all('a', class_='bookTitle')

p1_titles = [card.get_text(strip=True) for card in all_bookTitles]

p1_titles
```

```python
p1_titles_clean = [s.replace('\u200b', '') for s in p1_titles]
p1_titles_clean
```

### get urls for books

```python
all_urls = soup.find_all('a', attrs={'class': 'bookTitle', 'href': re.compile('^/book/show/')})

p1_urls = [card.get('href') for card in all_urls]

p1_urls
```

```python
string = 'https://www.goodreads.com'
p1_urls_clean = [string + s for s in p1_urls]
p1_urls_clean
```

### get number of times shelved as 'romantasy'

```python
shelved_romantasy = [t.find(string=re.compile("shelved")) for t in test]

shelved_romantasy
```

```python
shelved_romantasy = [i for i in shelved_romantasy if i]
    
shelved_romantasy
```

```python
shelved_romantasy = [re.sub("[^0-9]", "", x) for x in shelved_romantasy]
shelved_romantasy
```

### get authors for first page 

```python
all_p1_authors = soup.find_all('span', {'itemprop': 'name'})

p1_authors = [card.get_text(strip=True) for card in all_p1_authors]

p1_authors
```

### get book image for page 1

```python
# images = soup.find_all('a', attrs={'class':'leftAlignedImage', 'src': re.compile('^/https://i.gr-assets.com/images/S/compressed.photo.goodreads.com/books/')})
# image_srcs = [x.get('src') for x in images]
# image_srcs

images = [x.find('a', attrs={'class':'leftAlignedImage'}) for x in test]
images
```

```python
img_tag = [x.find('img') for x in images]
img_tag
```

```python
image_srcs = [x.get('src') for x in img_tag]
image_srcs
```

### get avg rating for page 1 

```python
meta = [t.find(string=re.compile('avg')) for t in test] 
#would prefer to split this into 3 lists but might have to add to df first 

meta
```

```python
meta = [i for i in meta if i]
    
meta
```

```python
#need to tidy the meta column a bit 
meta_cleaned1 = [x.replace('\n', '').replace('—', '').replace('                 ', '&').replace('avg rating','').replace('ratings','').replace('published','') for x in meta]
#created & to use as a delimiter later when i split the columns in the df 
meta_cleaned1
```

```python
meta_cleaned2 = [x.strip() for x in meta_cleaned1]
meta_cleaned2
```

## create df for first page 

```python
romantasy_page1_df = pd.DataFrame(
    {'Title': p1_titles,
     'Author': p1_authors,
     'Num_shelved_romantasy': shelved_romantasy,
     'Meta': meta_cleaned2,
     'Goodreads URL': p1_urls_clean,
     'Image URL': image_srcs
    })
```

```python
romantasy_page1_df.head()
```

```python
romantasy_page1_df.tail()
```

```python
meta_df = romantasy_page1_df['Meta'].str.split(pat='&', expand=True) #separate meta column into df to set up the columns 
```

```python
meta_df.columns = ['avg rating', 'num_ratings', 'year_published']
meta_df.head()
```

```python
romantasy_shelf_pg1 = romantasy_page1_df.join(meta_df)
romantasy_shelf_pg1.head()
```

```python
romantasy_shelf_pg1 = romantasy_shelf_pg1.drop('Meta', axis=1)
romantasy_shelf_pg1
```
