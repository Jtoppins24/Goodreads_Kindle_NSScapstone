```python
# Locate all spans that match the class for star ratings
star_spans = soup.find_all('span', class_='RatingStars RatingStars__small')

# List to store the star ratings
star_ratings = []

# Iterate through each span and extract the rating from `aria-label`
for span in star_spans:
    aria_label = span.get('aria-label', '')
    if 'Rating' in aria_label:
        # Extract the number of stars from 'Rating X out of 5'
        stars = int(aria_label.split()[1])  # Split the string into indiviidual words, extract the second word (the number), then use int() to turn it into an integer 
        star_ratings.append(stars)

print(star_ratings)
```

### get meta data for the reviewers - split into num reviews and num followers

```python
p1_meta = soup.find_all('div', class_='ReviewerProfile__meta')

p1_meta = [card.get_text() for card in p1_meta]

p1_meta
```

```python
len(p1_meta)
```

```python
meta_cleaned1 = [x.replace('reviews', '&').replace('followers','').strip() for x in p1_meta]
meta_cleaned1
#used & here so that I can split it later
```

## create df

```python
evelyn_hugo_reviews = pd.DataFrame(
    {'Reviewer': p1_reviewers,
     'Review Date': p1_dates,
     'Meta': meta_cleaned1,
     'Star Rating': star_ratings,
     'Review Text': p1_reviews
    })

evelyn_hugo_reviews.head()
```
